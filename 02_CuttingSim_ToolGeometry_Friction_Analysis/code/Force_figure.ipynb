```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path1='./cutting_force_vc20.txt'
path2 = './feed_force_vc20.txt'
path3='./cutting_force_vc40.txt'
path4='./feed_force_vc40.txt'
path5='./cutting_force_vc80.txt'
path6='./feed_force_vc80.txt'
data1= pd.read_table(path1, sep='\s+', header=None, names=['Time','Force'])
data2= pd.read_table(path2, sep='\s+', header=None, names=['Time','Force'])
data3= pd.read_table(path3, sep='\s+', header=None, names=['Time','Force'])
data4= pd.read_table(path4, sep='\s+', header=None, names=['Time','Force'])
data5= pd.read_table(path5, sep='\s+', header=None, names=['Time','Force'])
data6= pd.read_table(path6, sep='\s+', header=None, names=['Time','Force'])

def fix(data):
    data['Force'] *= 10**(-3)

data = [data1, data2, data3, data4, data5, data6]

for i in data:
    fix(i)

plt.figure(figsize=(8,16))
plt.subplot(311)
plt.plot(data1['Time'], data1['Force'], label='Cutting force')
plt.plot(data2['Time'], data2['Force'], label='Feed force')
plt.xlabel('Time(s)')
plt.ylabel('Force(N)')
plt.title('Force (vc=20)')
plt.legend()
plt.subplot(312)
plt.plot(data3['Time'], data3['Force'], label='Cutting force')
plt.plot(data4['Time'], data4['Force'], label='Feed force')
plt.xlabel('Time(s)')
plt.ylabel('Force(N)')
plt.title('Force (vc=40)')
plt.legend()
plt.subplot(313)
plt.plot(data5['Time'], data5['Force'], label='Cutting force')
plt.plot(data6['Time'], data6['Force'], label='Feed force')
plt.xlabel('Time(s)')
plt.ylabel('Force(N)')
plt.title('Force (vc=80)')

plt.legend()
plt.show()
```

```python
velocity = np.array([20, 40, 80])
exp_Fc = np.array([269, 234, 232])
sim_Fc = np.array([285.42, 259.83, 254.90])
my_Fc = np.array([569.53, 557.52, 537.16])

exp_Ff = np.array([235, 192, 181])
sim_Ff = np.array([128.09, 113.06, 104.74])
my_Ff = np.array([247.29, 270.24, 264.24])
```

```python
plt.plot(velocity, exp_Fc, 'C1--o', label='Exp_Fc')
plt.plot(velocity, sim_Fc, 'C2--o',label='Sim_Fc')
plt.plot(velocity, my_Fc, 'C3--o',label='My_Fc')

plt.plot(velocity, exp_Ff, 'C1-.x', label='Exp_Ff')
plt.plot(velocity, sim_Ff, 'C2-.x', label='Sim_Ff')
plt.plot(velocity, my_Ff, 'C3-.x', label='My_Ff')

plt.xlabel('Cutting velocity(m/min)')
plt.ylabel('Cutting and feed force(N)')
plt.legend(loc='upper center')
plt.show()
```
